# Data Preprocessing

## Drop ID, lat, long Columns

```python
trainX = trainX.drop(labels=['id','lat','long', 'zipcode','date'],axis=1)
print(trainX.columns)

testX = testX.drop(labels=['id','lat','long', 'zipcode','date'],axis=1)
testX.columns
```

## Column Type Conversion

```python
# Training Data

trainX['bathrooms'] = trainX['bathrooms'].astype('int64')
trainX['floors'] = trainX['floors'].astype('int64')
trainY['price'] = trainY['price'].astype('int64')
trainX.info()
```

```python
# Test Data

testX['bathrooms'] = testX['bathrooms'].astype('int64')
testX['floors'] = testX['floors'].astype('int64')
testX['waterfront'] = testX['waterfront'].astype('int64')
testY['price'] = testY['price'].astype('int64')
testX.info()
```

### One Hot Encoding

```python
# Taking a copy of Original data.
data1 = data.copy()

# Type Convertion.
data1['bathrooms'] = data1['bathrooms'].astype('object')
data1['floors'] = data1['floors'].astype('object')
data1['bedrooms'] = data1['bedrooms'].astype('object')
data1['waterfront'] = data1['waterfront'].astype('object')
data1['view'] = data1['view'].astype('object')
data1['condition'] = data1['condition'].astype('object')
data1['grade'] = data1['grade'].astype('object')                                               
data1['price'] = data1['price'].astype('int64')

# Dropping Columns
data1 = data1.drop(labels=['id','lat','long', 'zipcode','date'],axis=1)

# Dummies
data1_dummy = pd.get_dummies(data1)

X = data1_dummy.loc[:, data1_dummy.columns != 'price']
Y = data1_dummy[['price']]

from sklearn.model_selection import train_test_split

X_train1, X_val, Y_train1, Y_val = train_test_split(X, Y, random_state=10, test_size=0.1)
X_train, X_test, Y_train, Y_test = train_test_split(X_train1, Y_train1, random_state=10, test_size=0.2)

print(X_train.shape)        # 81 Columns


# Log Tranformation Of Price Variable.
Y_test['loged_price'] = np.log(Y_test['price'])     # Log Transform
Y_train['loged_price'] = np.log(Y_train['price'])     # Log Transform
```

## Dealing With Outliers

```python
# finding the 99th percentile Value trainY.

list1 = trainY.price.sort_values(ascending=True).tolist()
print(list1[15404])

# finding the 99th percentile Value testY.

list2 = testY.price.sort_values(ascending=True).tolist()
print(list2[3852])

# For One Hot Encoded Data.
list3 = Y_train.price.sort_values(ascending=True).tolist()
print(list3[15404])
```

```python
# Replacing Price Values above "1999000" with 1999000. [ with 99th percentile value. ]
# and storing in "price_outliers" variable.

trainY['price_outliers'] = pd.DataFrame(trainY.price.where(trainY['price'] < 1999000, 1999000))
print(trainY['price_outliers'].max())
print(trainY['price_outliers'].min())

# Replacing Price Values above "1950000" with 1950000. [ with 99th percentile value. ]
# and storing in price_outliers variable.
testY['price_outliers'] = pd.DataFrame(testY.price.where(testY['price'] < 1950000, 1950000))
print(testY['price_outliers'].max())
print(testY['price_outliers'].min())

# For One Hot Encoded Data.
Y_train['price_outliers'] = pd.DataFrame(Y_train.price.where(Y_train['price'] < 1999000, 1999000))
Y_test['price_outliers'] = pd.DataFrame(Y_test.price.where(Y_test['price'] < 1999000, 1999000))
```

# Model Building

## Linear Regression

### Linear Model Without Scaling/Normalizing Data. [Filled Outliers]

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error

import math

lm1 = LinearRegression()
lm1.fit(trainX, trainY['price_outliers'])

y_pred = lm1.predict(testX)

print("RMSE value ",math.sqrt(mean_squared_error(testY['price_outliers'] , y_pred)))
print("MSE value ",mean_squared_error(testY['price_outliers'] , y_pred))
print("MAE value ",mean_absolute_error(testY['price_outliers'] , y_pred))
print("R^2 score" ,r2_score(testY['price_outliers'] , y_pred))

lm1.coef_
```

## Regularization On Linear Model

### Lasso

```python
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.1)

clf.fit(trainX, trainY)
print(clf.coef_)

Y_cap  = clf.predict(testX)

print('\n')
print(math.sqrt(mean_squared_error(testY , Y_cap)))

print(r2_score(testY , Y_cap))
```
