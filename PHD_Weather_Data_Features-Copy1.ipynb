# Weather Data

## Imports And Data Loading

```python
import pandas as pd
from pandas import *
import numpy as np
```

```python
weather = pd.ExcelFile('C:\\Users\\HP\Desktop\\PHD Hackathon\\Data Science\\PhD_Train_Data\\WeatherDataNew.xlsx')
weather9 = pd.read_excel(weather, sheetname='2009')
weather10 = pd.read_excel(weather, sheetname='2010', na_values=[' ', '-'])
weather11 = pd.read_excel(weather, sheetname='2011', na_values=[' ', '-'])
weather12 = pd.read_excel(weather, sheetname='2012', na_values=[' ', '-'])
weather13 = pd.read_excel(weather, sheetname='2013', na_values=[' ', '-'])
weather14 = pd.read_excel(weather, sheetname='2014', na_values=[' ', '-'])
weather15 = pd.read_excel(weather, sheetname='2015', na_values=[' ', '-'])
weather16 = pd.read_excel(weather, sheetname='2016', na_values=[' ', '-'])
```

```python
df_list = [weather9, weather10, weather11, weather12, weather13, weather14, weather15, weather16]

weatherData = pd.concat(df_list)
weatherData.head()
```

```python
weatherData = weatherData.drop('Day',axis=1)
```

```python
weatherData.shape
```

```python
quant_list = [x*.1 for x in range(0,11)]
quant_list
```

```python
weatherData['Temp high (°C)'].quantile(quant_list)
```

```python
weatherData.columns = ['Year', 'Month', 'Temp high (°C)', 'Temp avg (°C)', 'Temp low (°C)',
                        'Dew Point high (°C)', 'Dew Point avg (°C)', 'Dew Point low (°C)',
                       'Humidity_high', 'Humidity_avg', 'Humidity_low','Sea Level Press high', 'Sea Level Press avg',
                       'Sea Level Press low', 'Visibility (km) high','Visibility (km) avg', 'Visibility (km) low', 'Wind (km/h) low',
                       'Wind (km/h) avg', 'Wind (km/h) high', 'Precip (mm) sum','WeatherEvent']
```

```python
weatherData['Precip (mm) sum'] =   weatherData['Precip (mm) sum'].apply(lambda x: 0.01 if x == 'T' else x)  
weatherData['Precip (mm) sum'].isnull().sum()

weatherData.iloc[:, 2:-1] = weatherData.iloc[:, 2:-1].apply(pd.to_numeric, errors = 'coerce')
```

```python
weatherData.dtypes
```

```python
bins = [0, 25, 50, 75, 100]
group_names = ['quant1','quant2','quant3','quant4']
#weatherDataCut = pd.DataFrame()

for i in weatherData.columns:
    if weatherData[i].dtype == 'float64':
        weatherData[i+'_bins'] = pd.cut(weatherData[i].values, bins, labels = group_names)
```

```python
weatherData.head()
```

```python
temp_high_pivot = pivot_table(weatherData, index=['Year','Month'], columns = ['Temp high (°C)_bins'],values = ['Temp high (°C)'])
temp_avg_pivot = pivot_table(weatherData, index=['Year','Month'], columns = ['Temp high (°C)_bins'],values = ['Temp high (°C)'])
temp_low_pivot = pivot_table(weatherData, index=['Year','Month'], columns = ['Temp high (°C)_bins'],values = ['Temp high (°C)'])

Dew_Point_high  = pivot_table(weatherData, index=['Year','Month'], columns = ['Dew Point high (°C)_bins'],values = ['Dew Point high (°C)'])
Dew_Point_avg  = pivot_table(weatherData, index=['Year','Month'], columns = ['Dew Point avg (°C)_bins'],values = ['Dew Point avg (°C)'])
Dew_Point_low  = pivot_table(weatherData, index=['Year','Month'], columns = ['Dew Point low (°C)_bins'],values = ['Dew Point low (°C)'])

Humidity_high = pivot_table(weatherData, index=['Year','Month'], columns = ['Humidity_high_bins'],values = ['Humidity_high'])
Humidity_avg = pivot_table(weatherData, index=['Year','Month'], columns = ['Humidity_avg_bins'],values = ['Humidity_avg'])
Humidity_low = pivot_table(weatherData, index=['Year','Month'], columns = ['Humidity_low_bins'],values = ['Humidity_low'])

Sea_Level_Press_high = pivot_table(weatherData, index=['Year','Month'], columns = ['Sea Level Press high_bins'],values = ['Sea Level Press high'])
Sea_Level_Press_avg = pivot_table(weatherData, index=['Year','Month'], columns = ['Sea Level Press avg_bins'],values = ['Sea Level Press avg'])
Sea_Level_Press_low = pivot_table(weatherData, index=['Year','Month'], columns = ['Sea Level Press low_bins'], values = ['Sea Level Press low'])

Visibility_high  = pivot_table(weatherData, index=['Year','Month'], columns = ['Visibility (km) high_bins'],values = ['Visibility (km) high'])
Visibility_avg  = pivot_table(weatherData, index=['Year','Month'], columns = ['Visibility (km) avg_bins'],values = ['Visibility (km) avg'])
Visibility_low  = pivot_table(weatherData, index=['Year','Month'], columns = ['Visibility (km) low_bins'],values = ['Visibility (km) low'])

Wind_high  = pivot_table(weatherData, index=['Year','Month'], columns = ['Wind (km/h) low_bins'],values = ['Wind (km/h) low'])
Wind_avg  = pivot_table(weatherData, index=['Year','Month'], columns = ['Wind (km/h) low_bins'],values = ['Wind (km/h) low'])
Wind_low  = pivot_table(weatherData, index=['Year','Month'], columns = ['Wind (km/h) low_bins'],values = ['Wind (km/h) low'])

Precip_sum  = pivot_table(weatherData, index=['Year','Month'], columns = ['Precip (mm) sum_bins'],values = ['Precip (mm) sum'])
```

```python
col = [temp_high_pivot, temp_avg_pivot, temp_low_pivot,Dew_Point_high,Dew_Point_avg,Dew_Point_low,Humidity_high,Humidity_avg,
 Humidity_low,Sea_Level_Press_high,Sea_Level_Press_avg,Sea_Level_Press_low,Visibility_high,Visibility_avg,Visibility_low,
Wind_high, Wind_avg, Wind_low, Precip_sum]
```
