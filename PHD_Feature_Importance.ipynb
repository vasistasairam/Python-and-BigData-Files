```python
trainX, valX, trainY, valY = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=True)
```

```python
rf_model = RandomForestRegressor(min_samples_leaf = 3, n_estimators=100, max_features='sqrt')
rf_model.fit(trainX, trainY)

print(rf_model.feature_importances_)
```

```python
X_cols = X.columns
feature_imp_RFC = pd.DataFrame({"imp_features":rf_model.feature_importances_, "columns":X_cols})
#feature_imp_RFC.plot.bar(x="columns", y="imp_features")
feature_imp_RFC.sort_values(by = 'imp_features', ascending=False)
```

### RF Prediction

### MAPE = [(abs(actual-predicted))/abs(actual)]*100

```python
y_pred = rf_model.predict(valX)

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(valY, y_pred)
```

### Selecting Only top 40% records

```python
feature_imp_RFC = feature_imp_RFC.sort_values('imp_features', ascending=False)
top_40_features = feature_imp_RFC.iloc[0:41, 0:1].values
```

```python
imp_columns = pd.DataFrame()
for i in top_40_features:
    for j in i:
        imp_columns[j] = X[j]
```
